## Regression Polinomial

Que podría pasar, si nuestra información no parece ser lineal??

```python
# librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data/estado1.csv")
df.head()
```

```python
X = np.arange(len(df["fecha"]))[0:30]
y = np.array(df["tmed"])[0:30]
```

```python
plt.figure(figsize=(30,5))
plt.scatter(X,y)
```

```python
# polinomio 4 potencia

plt.figure(figsize=(30,5))

p4 = np.poly1d(np.polyfit(X, y, 4))

plt.scatter(X, y)
plt.plot(X, p4(X), c="r")
plt.show()
```

```python
# polinomio 5 potencia

plt.figure(figsize=(30,5))

p5 = np.poly1d(np.polyfit(X, y, 5))

plt.scatter(X, y)
plt.plot(X, p5(X), c="r")
plt.show()
```

```python
# polinomio 6 potencia

plt.figure(figsize=(30,5))

p6 = np.poly1d(np.polyfit(X, y, 6))

plt.scatter(X, y)
plt.plot(X, p6(X), c="r")
plt.show()
```

```python
# polinomio 7 potencia

plt.figure(figsize=(30,5))

p7 = np.poly1d(np.polyfit(X, y, 7))

plt.scatter(X, y)
plt.plot(X, p7(X), c="r")
plt.show()
```

```python
# polinomio 8 potencia

plt.figure(figsize=(30,5))

p8 = np.poly1d(np.polyfit(X, y, 8))

plt.scatter(X, y)
plt.plot(X, p8(X), c="r")
plt.show()
```

```python
# polinomio 9 potencia

potencia = 9

plt.figure(figsize=(30,5))

p = np.poly1d(np.polyfit(X, y, potencia))

plt.scatter(X, y)
plt.plot(X, p(X), c="r")
plt.show()
```

```python
# polinomio 15 potencia

potencia = 15

plt.figure(figsize=(30,5))

p = np.poly1d(np.polyfit(X, y, potencia))

plt.scatter(X, y)
plt.plot(X, p(X), c="r")
plt.show()
```

## Evaluemos el valor de R2

```python
from sklearn.metrics import r2_score

r2 = r2_score(y,p(X))

print(r2)
```

## Automatizar evaluación

```python
plt.figure(figsize=(7,5))
results = []
potencias = range(10)

for potencia in potencias:

    p = np.poly1d(np.polyfit(X, y, potencia))
    r2 = r2_score(y,p(X))
    results.append(r2)
    
pd.Series(results, potencias).plot()
```
