## Tipo de distribuciones

```python
# librerías
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### Distribución Uniforme

```python
valores = np.random.uniform(-100.0, 100.0, 100000)
plt.hist(valores, 50)
plt.show()
```

### Normal o Gausiana

```python
# librerías
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```

Generar algunos numeros aleatoriamente con una distribución normal, 
* "mu" es la media deseada y 
* "sigma" es la desviación estandar

```python
mu = 5.0
sigma = 2.0
valores = np.random.normal(mu, sigma, 10000)
plt.hist(valores, 50)
plt.show()
```

## Exponential PDF / "Power Law"

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

## Binomial Probability Mass Function

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

# Poisson Probability Mass Function

Ejemplo, el portal del LNMySR tiene al día 250 visitas, Cual es la probabilidad de tener 300 visitas?

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 250
x = np.arange(150, 350, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```
